```
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/content/dataset2")
trainer.setTrainConfig(object_names_array=["no"], batch_size=4, num_experiments=50, train_from_pretrained_model="/content/pretrained-yolov3.h5")
trainer.trainModel()
```

```
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/content/dataset2")
trainer.evaluateModel(model_path="/content/dataset2/models", json_path="/content/dataset2/json/detection_config.json", iou_threshold=0.5, object_threshold=0.3, nms_threshold=0.5)
```

```
from imageai.Detection.Custom import CustomObjectDetection

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("/content/dataset2/models/detection_model-ex-037--loss-0005.129.h5") 
detector.setJsonPath("/content/dataset2/json/detection_config.json")
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image="/content/dataset2/validation/images/1011.jpg", output_image_path="madeiraNo4.jpg")
for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
```
